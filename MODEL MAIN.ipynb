```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)



import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
ragas = pd.read_csv('/Users/tishabhavsar/Downloads/Dataset.csv')

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(ragas.drop(columns=['Unnamed: 0','filename','raga']), ragas.raga, test_size=0.2, random_state=1)

from sklearn.preprocessing import MultiLabelBinarizer

multilabel_binarizer = MultiLabelBinarizer()
y_train = multilabel_binarizer.fit_transform(y_train)
y_test = multilabel_binarizer.fit_transform(y_test)

from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
sc_X_train = scale.fit_transform(X_train)
sc_X_test = scale.fit_transform(X_test)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(512, input_shape=(sc_X_train.shape[1],), activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(36, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
```

```python
#TRAINING CELL !!


model.fit(sc_X_train, y_train, epochs=100, batch_size=32)
```

```python
#SAVING CELL 

from tensorflow.keras.models import save_model
model.save('/Users/tishabhavsar/Library/Trained_model.keras')
```
